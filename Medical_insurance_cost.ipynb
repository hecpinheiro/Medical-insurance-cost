# **Leitura dos dados**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/hecpinheiro/Medical-insurance-cost-/main/insurance.csv')
dataset.head()
```


# **Pré-processamento dos dados**

```python
#Transformando os dados categóricos em númericos
dataset['smoker'] = dataset['smoker'].map({'no':0, 'yes': 1})
dataset['sex'] = dataset['sex'].map({'female':0, 'male':1})
dataset['region'] = dataset['region'].map({'southwest':0, 'southeast':1, 'northwest':2, 'northeast':3})

dataset.head()
```
